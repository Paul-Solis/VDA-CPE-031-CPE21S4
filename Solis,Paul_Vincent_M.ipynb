QUESTION 1:
 What is the overall prevalence of General Health in the dataset?

 The code visualizes the distribution of different General Health categories and their respective percentages within the dataset using a pie chart, providing insights into the overall prevalence of each category.

```python
import matplotlib.pyplot as plt

# Assuming 'HeartDisease' column with 'Yes' and 'No' values
heart_disease_counts = df['GeneralHealth'].value_counts()
labels = heart_disease_counts.index
sizes = heart_disease_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Overall Prevalence of Heart Disease')
plt.show
```

QUESTION #2: What is the average sleephours of individuals in the dataset?

The histogram reveals the distribution of sleep hours among individuals in the dataset, showing the frequency of different sleep durations. By observing the histogram, you can gain insights into the typical sleep patterns and identify any potential outliers or clusters of individuals with similar sleep habits.



```python
import matplotlib.pyplot as plt

plt.hist(df['SleepHours'], bins=10, edgecolor='black')  # Adjust bins as needed
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency')
plt.title('Distribution of Sleep Hours')
plt.show()
```
